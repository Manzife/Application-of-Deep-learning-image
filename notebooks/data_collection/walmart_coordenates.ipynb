The data is from https://www.kaggle.com/datasets/thedevastator/walmart-store-locations-across-the-united-states

```python
import pandas as pd
df = pd.read_csv("../Data/walmart_2018_11_06.csv")
df.head()
```

```python
# Ver si hay duplicados en una columna concreta
duplicados = df[df.duplicated(subset=["index"], keep=False)]
print(duplicados)

# Contar cuántos hay
print("Número de filas duplicadas:", df.duplicated(subset=["index"]).sum())
```

```python
df = df[["index", "state", "latitude", "longitude"]]
```

```python
df.rename(columns={"index": "store_id"}, inplace=True)
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
import pandas as pd

# Fijamos la cantidad exacta que quieres por estado
n_per_state = 2

# Obtenemos los índices aleatorios (hasta un máximo de 2 por estado, 
# por si algún estado en tu df tiene solo 1 Walmart)
indices_a_muestrear = (
    df.groupby("state", group_keys=False)
      .apply(lambda x: x.sample(min(len(x), n_per_state), random_state=42))
      .index
)

# Filtramos el DataFrame original usando esos índices
df_sample = df.loc[indices_a_muestrear]

print("Total de filas:", len(df_sample))
df_sample.head()
```

```python
df_sample["parquing"] = 0
```

```python
ids_con_parquing = df_sample["store_id"]

# 2. Usamos .isin() para filtrar y asignar el valor en una sola línea
df_sample.loc[df['store_id'].isin(ids_con_parquing), 'parquing'] = 1
```

```python
df_sample.head()
```

```python
df_sample.head(52)
```

```python
df_sample.loc[df_sample['store_id'] == 2801, 'parquing'] = 0
df_sample.loc[df_sample['store_id'] == 4445, 'parquing'] = 0
```

Both from DC don't have parquing, we have to find one that has it.

```python
df[df["state"] == "DC"].head(10)
```

The other one doesen't have a parquing nither, so we can't include any Walmart of DC.

```python
df_final = df_sample[df_sample["parquing"] == 1]
```

Let's now put the coordenates on the parquings, not the stores:

```python
df_final["parquing_coordenates"] = [[] for _ in range(len(df_final))]
```

```python
df_final.head(51)
```

```python
# We have applied this to all the stores_id
id_objetivo = 2729							
coordenadas = [45.829020, -108.468866]

idx = df_final.index[df_final['store_id'] == id_objetivo]
df_final.at[idx[0], 'parquing_coordenates'] = coordenadas
```
